```python
# specifing the column names that I want to use
new_cols_dict={
    'timestamp':'time',
    'cnt':'count',
    't1':'temp_real_C',
    't2':'temp_feels_like_C',
    'hum': 'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}
# Rename the column names
bike.rename(new_cols_dict,axis=1, inplace=True)
```

```python
# changing the humidity values to percentage (i.e, a value between o and 1)
bike.humidity_percent = bike.humidity_percent / 100
```

```python
# Create a season dictionary so that  we can map the integers 0-3 to actual written values
season_dict = {
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autumn',
    '3.0':'winter'
}

# Create a weather dictionary so that we can map the integer into actual written values
weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'}

# Changing the season column data type to string
bike.season = bike.season.astype('str')
# Mapping the values 0-3 to the actual written seasons
bike.season = bike.season.map(season_dict)

# Changing the wheather column data type to string
bike.wheather = bike.wheather.astype('str')
# Mapping the values to the actual written weathers
bike.wheather = bike.wheather.map(weather_dict)
```
